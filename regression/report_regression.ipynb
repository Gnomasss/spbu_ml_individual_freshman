
# **1. Генерация данных**


```
def f(x):
    yw = x ** 4 + x ** 2 + 1
    mu = np.mean(yw)
    sd = np.std(yw)
    jitter = 0.1 * np.random.normal(mu, sd, size=len(x))
    y = yw + jitter
    return y
```
